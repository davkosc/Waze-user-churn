```python
import numpy as np
import pandas as pd
laja_data = pd.read_excel(r'C:\Users\David.Kosciulek\Comp Documents\jupyter\laja_aggs\2022_laja_aggregation.xlsx',
                          sheet_name='Sheet1', header=0)
laja_data
```

```python
laja_data = laja_data.rename(columns={'F6.10b_C2 - Name of jurisdiction or landscape area':'Area'})
laja_data
```

```python
laja_area_pivot = pd.pivot_table(data=laja_data, index='Name of initiative', values='Area', aggfunc=', '.join)
laja_area_pivot.to_excel(r'C:\Users\David.Kosciulek\Comp Documents\jupyter\laja_aggs\laja_area_pivot.xlsx')
print(laja_area_pivot)
```

```python
laja_confidence_pivot = pd.pivot_table(data=laja_data, index='Name of initiative', values='Confidence', aggfunc=', '.join)
laja_confidence_pivot.to_excel(r'C:\Users\David.Kosciulek\Comp Documents\jupyter\laja_aggs\laja_confidence_pivot.xlsx')
print(laja_confidence_pivot)
```
